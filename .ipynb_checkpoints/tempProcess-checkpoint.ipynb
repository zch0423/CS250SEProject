```python
import pandas as pd
```

```python
pd.read_json(issues)
```

```python
data = pd.DataFrame(issues)
```

```python
data = data[["number", "title", "body", "comments", "labels", "created_at"]]
```

```python
data
```

```python
labels = data["labels"]
```

```python
labels
```

```python
count = 0
for each in labels:
    count+=1
print(count)
```

```python
tempLabels = []
for row in labels:
    temp = []
    if row:
        for each in row:
            temp.append(each["name"])
    tempLabels.append(temp)
```

```python
len(tempLabels)
```

```python
len([1 for each in tempLabels if each])
```

```python
tempLabels
```

```python
data["labels"] = tempLabels
```

```python
data.loc["labels"] = tempLabels
```

```python
data.loc[:, ('labels')] = tempLabels
```
