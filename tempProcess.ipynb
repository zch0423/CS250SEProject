```python
def startWithType(x):
    for each in x:
        if each.startswith("type"):
            return True
    return False
def startWithTheme(x):
    for each in x:
        if each.startswith("theme"):
            return True
    return False
def startWithComp(x):
    for each in x:
        if each.startswith("com"):
            return True
    return False
```

```python
labelsWithType = data[data["labels"].apply(lambda x:startWithType(x))]
```

```python
labelsWithType
```

```python
temp = []
for each in labelsWithType["labels"]:
    for x in each:
        if x.startswith("type"):
            temp.append(x)
            break
```

```python
temp
```

```python
labelsWithType.loc[:, ("labels")] = temp
```
